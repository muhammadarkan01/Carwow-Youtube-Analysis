# **2. Sentiment Labelling**

```python
!pip install transformers torch
```

```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch
```

```python
from tqdm import tqdm
```

```python
# Instantiate pre-trained model
tokenizer = AutoTokenizer.from_pretrained('nlptown/bert-base-multilingual-uncased-sentiment')
model = AutoModelForSequenceClassification.from_pretrained('nlptown/bert-base-multilingual-uncased-sentiment')
```

```python
# Define function to execute the pre-trained model
def sentiment_score(text):
    tokens = tokenizer.encode(text, return_tensors='pt')
    result = model(tokens)
    return int(torch.argmax(result.logits))+1
```

```python
# # Slicing dataset
# new_df = df_comments[:6000]
# new_df.info()
# new_df.head()
```

```python
# # Execute sentiment labelling
# tqdm.pandas(desc='My bar!')
# new_df['sentiment_score'] = new_df['preprocessed_text'].progress_apply(lambda x: sentiment_score(x))

# Execute sentiment labelling
tqdm.pandas(desc='My bar!')
df_comments['sentiment_score'] = df_comments['updated_text'].progress_apply(lambda x: sentiment_score(x[:512]))
```
