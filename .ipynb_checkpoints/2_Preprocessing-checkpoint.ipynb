# **2. Preprocessing Function**

```python
# Load the stopwords
stopwords = nltk.corpus.stopwords.words('english')

# Define a function to preprocess the text
def preprocess_text(text):
    # Convert the text to lowercase
    text = text.lower()

     # Remove URLs
    text = re.sub(r'http\S+', '', str(text))

    # Remove usernames
    text = re.sub(r'@\w+', '', str(text))

    # Remove hashtags
    text = re.sub(r'#\w+', '', str(text))

    # Remove punctuation and special characters
    text = re.sub(r'[^\w\s]', '', str(text))

    # Remove emoticons
    text = re.sub(r':\)|;\)|:-\)|\(-:|:\'\(|:\(|:-\(|:O|8\)|:\*|>:O|:\-\/|\^\^|:D|:P|:S|:\|', '', str(text))

    # Remove extra whitespace
    text = re.sub(r'\s+', ' ', str(text)).strip()

    # Remove stopwords
    text = [word for word in text.split() if word not in stopwords]

    # Detokenize the tokens to form text
    text = ' '.join(text)

    return text
```

# **3. Preprocessing Comments**

```python
# Apply the preprocess_text() function on text column
df_comment['preprocessed_text'] = df_comment['text'].apply(lambda x: preprocess_text(x))
df_comment.info()
df_comment.head()
```

# **4. Preprocessing Videos**

```python
# Apply the preprocess_text() function on title column
df_videos['preprocessed_title'] = df_videos['title'].apply(lambda x: preprocess_text(x))
df_videos.info()
df_videos.head()
```

```python
# Change data type of upload_date to datetime
df_videos['upload_date'] = pd.to_datetime(df_videos['upload_date'], format='%Y%m%d')
df_videos.info()
```

```python
# Apply the preprocess_text() function on title column
df_videos['preprocessed_tags'] = df_videos['tags'].apply(lambda x: preprocess_text(x))
df_videos.info()
df_videos.head()
```

# **5. Labelling Videos (Brands)**

```python
from sklearn.feature_extraction.text import CountVectorizer

def find_most_frequent_unigram(df, column):
    # Initialize the CountVectorizer
    vectorizer = CountVectorizer()

    # Fit and transform the specified text column
    unigram_features = vectorizer.fit_transform(df[column])

    # Convert the resulting sparse matrix to a DataFrame
    unigram_df = pd.DataFrame(unigram_features.toarray(), columns=vectorizer.get_feature_names_out())

    # Find the top 100 most frequent unigrams
    top_unigrams = unigram_df.sum().nlargest(100).index.tolist()

    # Get the corresponding counts for the top unigrams
    counts = unigram_df[top_unigrams].sum().tolist()

    return top_unigrams, counts
```

```python
def find_top_bigrams_with_counts(df, column):
    # Initialize the CountVectorizer for bigrams
    vectorizer = CountVectorizer(ngram_range=(2, 2))

    # Fit and transform the specified text column
    bigram_features = vectorizer.fit_transform(df[column])

    # Convert the resulting sparse matrix to a DataFrame
    bigram_df = pd.DataFrame(bigram_features.toarray(), columns=vectorizer.get_feature_names_out())

    # Calculate the sum of each bigram and get the top 100
    top_bigrams = bigram_df.sum().nlargest(100).index.tolist()

    # Get the corresponding counts for the top bigrams
    counts = bigram_df[top_bigrams].sum().tolist()

    return top_bigrams, counts
```

```python
search = 'nissan'
df_videos[df_videos['preprocessed_title'].str.contains(rf'\b{search}\b', case=False, regex=True)]
```

```python
tags_unigram, counts = find_most_frequent_unigram(df_videos, 'preprocessed_tags')

# Display the results
for unigram, count in zip(tags_unigram, counts):
    print(f"{unigram}: {count} occurrences")
```

```python
tags_bigram, counts = find_top_bigrams_with_counts(df_videos, 'preprocessed_tags')

# Display the results
for bigram, count in zip(tags_bigram, counts):
    print(f"{bigram}: {count} occurrences")
```

```python
# Create a list of car brands based on the unigram and bigram results
car_brands_list = [
    'porsche', 'ferrari', 'lamborghini', 'bmw', 'mercedes', 'mclaren', 'bentley',
    'nissan', 'lexus', 'rollsroyce', 'audi', 'skoda', 'volvo', 'land rover', 'ford',
    'lucid', 'brabus', 'dodge', 'tesla', 'vw', 'bugatti', 'toyota', 'honda',
    'peugeot', 'ds' , 'corvette', 'suzuki', 'mitsubishi', 'range rover',
    'genesis', 'rimac', 'aston martin', 'caterham', 'mazda',
    'citroen', 'lotus', 'mg', 'gordon murray automotive', 'human horizons',
    'byd', 'smart'
]

# List down the car brands in each video
def extract_car_brands(title, car_brands_list):
    car_brands = [brand for brand in car_brands_list if re.search(rf'\b{re.escape(brand)}\b', title)]
    return car_brands

# Apply the function to create a new column 'car_brands'
df_videos['car_brands'] = df_videos['preprocessed_tags'].apply(lambda x: extract_car_brands(x, car_brands_list))
df_car_brands = df_videos['car_brands'].value_counts().reset_index()
df_car_brands.columns = ['brands', 'count']
df_car_brands
```

```python
# List to check for in 'car_brands'
brand_list_to_check = ['mg']

# Filter rows where 'car_brands' contains the specified list
df_videos[df_videos['car_brands'].apply(lambda x: set(brand_list_to_check).issubset(x))]
```

# **6. Labelling Video Categories (Race, Review, News, Comparison)**

```python
# Function to remove elements in car_brands_list
def remove_car_brands(s):
    words = s.split()
    filtered_words = [word for word in words if word.lower() not in car_brands_list]
    return ' '.join(filtered_words)

df_videos['updated_title'] = df_videos['preprocessed_title'].apply(remove_car_brands)
df_videos.info()
df_videos.head()
```

```python
title_unigram, counts = find_most_frequent_unigram(df_videos, 'updated_title')

# Display the results
for unigram, count in zip(title_unigram, counts):
    print(f"{unigram}: {count} occurrences")
```

```python
title_bigram, counts = find_top_bigrams_with_counts(df_videos, 'updated_title')

# Display the results
for bigram, count in zip(title_bigram, counts):
    print(f"{bigram}: {count} occurrences")
```

```python
# Define keyword mappings
keyword_mappings = {
    'races': {'drag', 'race', 'beat'},
    'review': {'review', 'reviewed', 'drove', 'budget'},
    'news': {'revealed', 'new'},
    'comparison': {'best'}
}

# Function to classify the video_category
def classify_category(title):
    if pd.notna(title):  # Check if the value is not NaN
        for classification, keywords in keyword_mappings.items():
            if any(re.search(rf'\b{re.escape(keyword)}\b', title) for keyword in keywords):
                return classification
        return 'other'

# Apply the function to classify video_category
df_videos['video_category'] = df_videos['updated_title'].apply(classify_category)
df_videos['video_category'].value_counts()
```
