# **1. File Conversion**

```python
import pandas as pd
from google.colab import drive
drive.mount('/content/drive')
```

```python
# Create dataframe for videos
video_path = '/content/drive/MyDrive/Carwow-Youtube-Analysis/Notebook - Copy/df_videos_combined.csv'
df_videos = pd.read_csv(video_path)
df_videos.info()
df_videos
```

```python
df_videos.drop_duplicates(subset=['id', 'title'], keep='first', inplace=True)
df_videos.info()
df_videos
```

```python
df_videos.to_excel('carwow_videos_combined.xlsx', index=False)
```

```python
# Create dataframe for comments
comment_path = '/content/drive/MyDrive/Carwow-Youtube-Analysis/Notebook - Copy/df_comments_combined.csv'
df_comments = pd.read_csv(comment_path)
df_comments.info()
df_comments
```

```python
df_comments.drop_duplicates(subset=['id'], keep='first', inplace=True)
df_comments.info()
df_comments
```

```python
df_comments.to_excel('carwow_comments_combined.xlsx', index=False)
```
