# Preprocess

## Join vids month upload with comments

```python
import re

def extract_main_id(input_string):
    match = re.match(r'[^\w]*([\w]+)\.[^\w]*', input_string)
    return match.group(1) if match else input_string
```

```python
df_videos['upload_date'] = pd.to_datetime(df_videos['upload_date'], format='%Y-%m-%d')
df_videos
```

```python
# Assuming 'df' is your DataFrame and 'date_column' is the datetime column
df_videos['upload_month'] = df_videos['upload_date'].dt.month

# Display the DataFrame with the extracted month
df_videos
```

```python
# Select the columns you want from df_videos
columns_to_join = ['id', 'upload_month']

# Perform the join on selected columns
result_df = pd.merge(df_comments, df_videos[columns_to_join], left_on='video_id', right_on='id', how='inner')

result_df
```

# Cohort Analysis

```python
df_cohort = result_df[['video_id','id_y','author','upload_month']]
df_cohort
```

```python
df_cohort['author'].value_counts()
```

```python
df_cohort['cohort_month'] = df_cohort.groupby('author')['upload_month'].transform(min)
df_cohort
```

```python
df_cohort.drop(columns=['video_id', 'id_y'], inplace=True)
```

```python
df_cohort
```

```python
df_cohort['cohort_index'] = df_cohort['upload_month'] - df_cohort['cohort_month'] + 1
df_cohort
```

```python
df_cohort.sort_values(by='cohort_index', ascending=False)
```

```python
cohort_data = df_cohort.groupby(['cohort_month', 'cohort_index'])['author'].nunique().reset_index()
cohort_data
```

```python
cohort_table = cohort_data.pivot(index='cohort_month', columns=['cohort_index'], values='author')
cohort_table
```

```python
cohort_table_1 = cohort_table.divide(cohort_table.iloc[:, 0], axis=0)
cohort_table_1
```

```python
cohort_table_2 = cohort_table_1.copy()
```

```python
cohort_table_2
```

```python
# Mapping function to convert numeric month to month name with "2023"
def convert_to_month_name(month_numeric):
    month_names = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
    month_name = month_names[month_numeric - 1]  # Adjust for 0-based index
    return f"{month_name} 2023"

# Apply the mapping function to convert cohort_index to month names
cohort_table_2.index = cohort_table_2.index.map(convert_to_month_name)
```

```python
cohort_table_2
```

# Table mapping

```python
import seaborn as sns
import matplotlib.pyplot as plt

heatmap_data = cohort_table_2

# Create a heatmap
plt.figure(figsize=(15, 10))  # Set the size of the plot
sns.heatmap(heatmap_data, cmap='Blues', annot=True, fmt=".0%", linewidths=.5)

# Add labels and a title
plt.title('Youtube Commenter Cohort Analysis')
# plt.xlabel('Your_Columns')
# plt.ylabel('Your_Index')

# Show the plot
plt.show()
```
