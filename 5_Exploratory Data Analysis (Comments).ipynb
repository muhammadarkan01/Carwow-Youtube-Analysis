## **3.2 Sentiment Distribution**

```python
import matplotlib.pyplot as plt

# Sentiment Distribution
def display_sentiment_dist(df):

 # Count the occurrences of each sentiment label
  sentiment_counts = df['sentiment_label'].value_counts()

  # Define colors for each sentiment label
  colors = {'positive': 'green', 'negative': 'red', 'neutral': 'gray'}

  # Plotting the doughnut chart
  fig, ax = plt.subplots(figsize=(8, 8))

  # Plot doughnut chart
  wedges, texts, autotexts = ax.pie(sentiment_counts, labels=sentiment_counts.index,
                                    autopct=lambda p: '{:.1f}%\n({:.0f})'.format(p, p/100*sum(sentiment_counts)),
                                    startangle=140, wedgeprops=dict(width=0.4, edgecolor='w'), colors=[colors[label] for label in sentiment_counts.index],
                                    pctdistance=0.85)  # Adjust the percentage distance

  # Set text color inside the doughnut chart to white for better visibility
  for autotext in autotexts:
      autotext.set_color('white')

  # Add legend with counts for each sentiment label
  legend_labels = [f'{label} ({count})' for label, count in zip(sentiment_counts.index, sentiment_counts)]
  ax.legend(wedges, legend_labels, title='Sentiment', loc='center left', bbox_to_anchor=(1, 0, 0.5, 1))

  # Add a title and total count
  ax.set_title(f'Sentiment Distribution of Comments in Carwow YouTube Channel')

  # Display the plot
  plt.show()
```

```python
display_sentiment_dist(df_comments)
```

## **3.3 Wordcloud Unigram and Bigram**

```python
from wordcloud import WordCloud
from sklearn.feature_extraction.text import CountVectorizer

# Wordcloud (Unigram)
def generate_sentiment_word_clouds(df, ngram):
    def generate_word_cloud(text, title, color='viridis'):
        wordcloud = WordCloud(width=800, height=400, background_color='white', colormap=color).generate(text)

        plt.figure(figsize=(8, 4))
        plt.imshow(wordcloud, interpolation='bilinear')
        plt.title(title)
        plt.axis('off')
        plt.show()

    # Convert 'updated_text' column to strings to handle potential float values
    df['updated_text'] = df['updated_text'].astype(str)

    # Function to generate bigrams
    def generate_bigrams(text):
        vectorizer = CountVectorizer(ngram_range=(2, 2), stop_words='english')
        bigrams = vectorizer.fit_transform([text])
        feature_names = vectorizer.get_feature_names_out()
        return ' '.join(feature_names)

    if ngram == 'uni':
      positive_text = ' '.join(df[df['sentiment_label'] == 'positive']['updated_text'])
      neutral_text = ' '.join(df[df['sentiment_label'] == 'neutral']['updated_text'])
      negative_text = ' '.join(df[df['sentiment_label'] == 'negative']['updated_text'])

    if ngram == 'bi':
      positive_text = generate_bigrams(' '.join(df[df['sentiment_label'] == 'positive']['updated_text']))
      neutral_text = generate_bigrams(' '.join(df[df['sentiment_label'] == 'neutral']['updated_text']))
      negative_text = generate_bigrams(' '.join(df[df['sentiment_label'] == 'negative']['updated_text']))

    generate_word_cloud(positive_text, 'Positive Comments on Carwow YouTube Channel', 'Greens')
    generate_word_cloud(neutral_text, 'Neutral Comments on Carwow YouTube Channel', 'Greys')
    generate_word_cloud(negative_text, 'Negative Comments on Carwow YouTube Channel', 'Reds')
```

```python
# Wordcloud (Unigram)
generate_sentiment_word_clouds(df_comments, 'uni')
```

```python
# Wordcloud (Bigram)
generate_sentiment_word_clouds(df_comments, 'bi')
```

```python
from collections import Counter
# Function to extract and count bigrams
def count_bigrams(text_series, n_most_common=20):
    # Use CountVectorizer to extract bigrams
    vectorizer = CountVectorizer(ngram_range=(2, 2), stop_words='english')
    bigrams = vectorizer.fit_transform(text_series)

    # Get feature names (bigrams) and their counts
    feature_names = vectorizer.get_feature_names_out()
    bigram_counts = bigrams.sum(axis=0).A1

    # Create a dictionary of bigrams and their counts
    bigram_counter = dict(zip(feature_names, bigram_counts))

    # Get the most common bigrams
    most_common_bigrams = Counter(bigram_counter).most_common(n_most_common)

    return most_common_bigrams
```

```python
# Positive Comment
# Count bigrams in the 'updated_text'
most_common_bigrams = count_bigrams(df_comments[df_comments['sentiment_label']=='positive']['updated_text'].astype(str))

# Display the most common bigrams and their counts
for bigram, count in most_common_bigrams:
    print(f'{bigram}: {count} occurrences')
```

```python
# Neutral Comment
# Count bigrams in the 'updated_text'
most_common_bigrams = count_bigrams(df_comments[df_comments['sentiment_label']=='neutral']['updated_text'].astype(str))

# Display the most common bigrams and their counts
for bigram, count in most_common_bigrams:
    print(f'{bigram}: {count} occurrences')
```

```python
# Negative Comment
# Count bigrams in the 'updated_text'
most_common_bigrams = count_bigrams(df_comments[df_comments['sentiment_label']=='negative']['updated_text'].astype(str))

# Display the most common bigrams and their counts
for bigram, count in most_common_bigrams:
    print(f'{bigram}: {count} occurrences')
```

```python
# Assuming df is your DataFrame
corpus = filtered_df['preprocessed_text']
sentiment_labels = filtered_df['sentiment_label']

# Unigram analysis grouped by sentiment label
unigram_vectorizer = CountVectorizer(stop_words='english', max_features=10)
unigram_matrix = unigram_vectorizer.fit_transform(corpus)
unigram_features = unigram_vectorizer.get_feature_names_out()
unigram_counts = pd.DataFrame(unigram_matrix.toarray(), columns=unigram_features)
unigram_counts['sentiment_label'] = sentiment_labels

# Bigram analysis grouped by sentiment label
bigram_vectorizer = CountVectorizer(ngram_range=(2, 2), stop_words='english', max_features=10)
bigram_matrix = bigram_vectorizer.fit_transform(corpus)
bigram_features = bigram_vectorizer.get_feature_names_out()
bigram_counts = pd.DataFrame(bigram_matrix.toarray(), columns=bigram_features)
bigram_counts['sentiment_label'] = sentiment_labels

# Display results
print("Top 10 Unigrams Grouped by Sentiment Label:")
print(unigram_counts.groupby('sentiment_label').sum().T)

print("\nTop 10 Bigrams Grouped by Sentiment Label:")
print(bigram_counts.groupby('sentiment_label').sum().T)
```
