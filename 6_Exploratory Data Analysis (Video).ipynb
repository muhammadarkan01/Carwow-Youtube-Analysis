## **2.3 Upload Trend**

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_videos['upload_date'] = pd.to_datetime(df_videos['upload_date'], format='%Y-%m-%d')
df_videos['upload_month'] = df_videos['upload_date'].dt.to_period('M')
monthly_counts = df_videos['upload_month'].value_counts().sort_index()

# Normalize the counts to the range [0, 1]
normalized_counts = (monthly_counts - monthly_counts.min()) / (monthly_counts.max() - monthly_counts.min())

# Create a gradient within a single color (blue)
blue_palette = sns.color_palette("RdBu", len(monthly_counts))
colors = sns.color_palette("RdBu", len(monthly_counts), as_cmap=True)(normalized_counts)

# Create bar chart with gradient color
plt.figure(figsize=(15, 7))
sns.barplot(x=monthly_counts.index.astype(str), y=monthly_counts.values, palette=colors)
plt.title('Monthly Trend of Video Uploads by Carwow')
plt.xlabel('Month')
plt.ylabel('Number of Uploads')
plt.xticks(rotation=90)
plt.show()
```

## **2.4 Video Category Distribution**

```python
# Count occurrences of each category
category_counts = df_videos['video_category'].value_counts()

# Calculate percentages
category_percentages = (category_counts / len(df_videos)) * 100

# Plot the pie chart
fig, ax = plt.subplots()

# Plot the pie chart with percentages only
wedges, texts, autotexts = ax.pie(category_counts, labels=None, autopct='%1.1f%%', startangle=90, colors=plt.cm.Set3.colors)

# Add category labels and percentages within each pie chart area
for autotext, (category, percentage) in zip(autotexts, zip(category_counts.index, category_percentages)):
    autotext.set_text(f'{percentage:.1f}%')
    autotext.set_fontsize(7)  # Adjust text size

# Add legend on the edge of the plot
ax.legend(wedges, category_counts.index, title='Categories', loc='center left', bbox_to_anchor=(1, 0.5), fontsize=8)

# Add total on the edge of the plot
total_text = f'Total: {len(df_videos)}'
ax.text(1.2, 0.5, total_text, va='center', ha='center', fontsize=10, color='black')

# Set plot title
plt.title('Video Category Distribution')

# Equal aspect ratio ensures that the pie is drawn as a circle.
ax.axis('equal')

# Show the plot
plt.show()
```

## **2.5 Top Brands**

```python
import ast

# Convert string representations of lists to actual lists
df_videos['car_brands'] = df_videos['car_brands'].apply(ast.literal_eval)

# Create a flat list
flat_list = [item for sublist in df_videos['car_brands'] for item in sublist]

print(flat_list)
```

```python
from collections import Counter

# Count occurrences of each brand
brand_counts = Counter(flat_list)

# Sort the counts in descending order
sorted_brand_counts = dict(sorted(brand_counts.items(), key=lambda x: x[1], reverse=True))

# Display the sorted counts
for brand, count in sorted_brand_counts.items():
    print(f'{brand}: {count}')
```

```python
import numpy as np

# Count occurrences of each brand
brand_counts = Counter(flat_list)

# Sort the counts in descending order
sorted_brand_counts = dict(sorted(brand_counts.items(), key=lambda x: x[1], reverse=True))

# Extract data for plotting
brands = list(sorted_brand_counts.keys())
counts = list(sorted_brand_counts.values())

# Create a gradient colormap based on counts
colors = np.array(counts)
norm = plt.Normalize(colors.min(), colors.max())
cmap = plt.cm.get_cmap('Blues')

# Create a bar chart with gradient colors
plt.figure(figsize=(20, 10))
bars = plt.bar(brands, counts, color=cmap(norm(colors)))

# Add a colorbar for reference
sm = plt.cm.ScalarMappable(cmap=cmap, norm=norm)
sm.set_array([])
cbar = plt.colorbar(sm, orientation='vertical')
cbar.set_label('Count Gradient')

# Add labels and a title
plt.title('Brand Counts Bar Chart')
plt.xlabel('Brand')
plt.ylabel('Count')

# Rotate x-axis labels for better readability (optional)
plt.xticks(rotation=45, ha='right')

# Show the plot
plt.show()
```

## **2.6 Trends Performance**


```python
df_videos['upload_month'] = df_videos['upload_date'].dt.month
df_videos
```

```python
# Assuming df is your DataFrame with the 'upload_month' column
df_result = pd.DataFrame(df_videos['upload_month'].value_counts().reset_index())
df_result.columns = ['Month', 'Number of vids']

# Sort the DataFrame by the 'Month' column in ascending order
df_result = df_result.sort_values(by='Month')

df_result
```

```python
df_result_2 = df_videos.groupby('upload_month')['view_count'].mean().reset_index()
df_result_2.columns = ['Month', 'Avg view_count']

df_result_2
```

```python
df_result_3 = df_videos.groupby('upload_month')['comment_count'].mean().reset_index()
df_result_3.columns = ['Month', 'Avg comment_count']

df_result_3
```

```python
df_result_4 = df_videos.groupby('upload_month')['like_count'].mean().reset_index()
df_result_4.columns = ['Month', 'Avg like_count']

df_result_4
```

```python
merged_df = pd.merge(df_result, df_result_2, on='Month', how='inner')
merged_df = pd.merge(merged_df, df_result_3, on='Month', how='inner')
merged_df = pd.merge(merged_df, df_result_4, on='Month', how='inner')

merged_df
```

```python
# import pandas as pd
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Extract the 'Month' column for the x-axis
months = merged_df['Month']

# Extract other columns for scaling
features_to_scale = merged_df.columns[2:]

# Apply Min-Max scaling to the features
scaler = MinMaxScaler()
df_scaled = pd.DataFrame(scaler.fit_transform(merged_df[features_to_scale]), columns=features_to_scale)

# Plot the line chart
fig, ax1 = plt.subplots(figsize=(15, 10))

# Plot Number of videos on the first axis
ax1.bar(months, merged_df['Number of vids'], color='c', label='Number of videos', alpha=0.5, align='center')
ax1.set_xlabel('Month')
ax1.set_ylabel('Number of videos', color='k')
ax1.tick_params('y', colors='k')

# Create a second y-axis for the scaled averages
ax2 = ax1.twinx()
lines = ax2.plot(months, df_scaled, linestyle='-', marker='o')
ax2.set_ylabel('Averages (Scaled)', color='k')
ax2.tick_params('y', colors='k')

# # Set custom colors for the lines in the second axis
# colors = plt.get_cmap('cividis')(np.linspace(0, 1, len(lines)))
# for i, line in enumerate(lines):
#     line.set_color(colors[i])

# Use different line styles
line_styles = ['--', ':', '-.']
for i, line in enumerate(lines):
    line.set_linestyle(line_styles[i % len(line_styles)])

# Add markers
marker_styles = ['v', '>', '*', 's', '^', 'p']
for i, line in enumerate(lines):
    line.set_marker(marker_styles[i % len(marker_styles)])

# Display the legend for the scaled averages
labels = [f'Scaled {feature}' for feature in features_to_scale]
ax2.legend(lines, labels, loc='upper right')

# Display the legend for both axes
ax1.legend(loc='upper left')

# Show the plot
plt.title('Line Chart with Three Axes and Legends')
plt.show()
```
